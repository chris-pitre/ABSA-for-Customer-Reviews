```python
# Function to analyze a single sentence
def analyze_sentence(sentence):
    return aspect_extractor.extract_aspect(
        inference_source=[sentence],
        pred_sentiment=True,
    )
```

```python
import pandas as pd
import nltk
from pyabsa import available_checkpoints, ATEPCCheckpointManager


nltk.download('punkt')
from nltk.tokenize import sent_tokenize


checkpoint_map = available_checkpoints()


aspect_extractor = ATEPCCheckpointManager.get_aspect_extractor(
    checkpoint='english',
    auto_device=True
)


review_text = "The aftertaste is just right – not too sweet, not too bitter. However, the high calorie and sugar content is becoming a bigger issue for me as I try to adopt a healthier lifestyle."


sentences = nltk.tokenize.sent_tokenize(review_text)

atepc_result = []
for sentence in sentences:

    atepc_result.append(aspect_extractor.extract_aspect(
        inference_source=[sentence],
        pred_sentiment=True,
    )[0])
print(atepc_result)
```

```python
import pandas as pd

df = pd.DataFrame(atepc_result)

df
```

```python

good_features = []
bad_features = []


for sentence in sentences:
    sentence_result = analyze_sentence(sentence)
    for result in sentence_result:
        for aspect in result['aspect']:
            aspect_term = aspect['term']
            sentiment = aspect['sentiment']

            if sentiment == 'positive':
                good_features.append(aspect_term)
            elif sentiment == 'negative':
                bad_features.append(aspect_term)


df = pd.DataFrame({
    'Good Features': pd.Series(good_features),
    'Bad Features': pd.Series(bad_features)
})

print(df)
```
